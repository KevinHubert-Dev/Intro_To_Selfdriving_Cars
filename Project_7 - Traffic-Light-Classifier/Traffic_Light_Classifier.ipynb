# 2. Pre-process the Data

After loading in each image, you have to standardize the input and output!

### Input

This means that every input image should be in the same format, of the same size, and so on. We'll be creating features by performing the same analysis on every picture, and for a classification task like this, it's important that **similar images create similar features**! 

### Output

We also need the output to be a label that is easy to read and easy to compare with other labels. It is good practice to convert categorical data like "red" and "green" to numerical data.

A very common classification output is a 1D list that is the length of the number of classes - three in the case of red, yellow, and green lights - with the values 0 or 1 indicating which class a certain image is. For example, since we have three classes (red, yellow, and green), we can make a list with the order: [red value, yellow value, green value]. In general, order does not matter, we choose the order [red value, yellow value, green value] in this case to reflect the position of each light in descending vertical order.

A red light should have the  label: [1, 0, 0]. Yellow should be: [0, 1, 0]. Green should be: [0, 0, 1]. These labels are called **one-hot encoded labels**.

*(Note: one-hot encoding will be especially important when you work with [machine learning algorithms](https://machinelearningmastery.com/how-to-one-hot-encode-sequence-data-in-python/)).*

### (IMPLEMENTATION): Standardize the input images

* Resize each image to the desired input size: 32x32px.
* (Optional) You may choose to crop, shift, or rotate the images in this step as well.

It's very common to have square input sizes that can be rotated (and remain the same size), and analyzed in smaller, square patches. It's also important to make all your images the same size so that they can be sent through the same pipeline of classification steps!

````python
# TODO: Resize image and pre-process so that all "standard" images are the same size  
# This function should take in an RGB image and return a new, standardized version
def standardize_input(image):
    # Cause i won't rotate the image i don't make them square
    img_height = image.shape[0]
    img_width = image.shape[1]
    crop_y = round(img_height * 0.1) # 0.1 = 10%
    
    crop_x = round(img_height * 0.1)
    
    # Crop 10 percent of top, left, right and bottom of image.
    # The cropped parts of the provided images are mostly objects (e.g. tree, sky) around the traffic-light
    # which are less interesting for the classification.
    crop_img = image[crop_y:img_height - crop_y,crop_x:img_width - crop_x]
    
    # New height, width for the image
    newWidth = 32
    newHeight = 32
    
    # Resize image 
    result_img = cv2.resize(crop_img, (newWidth, newHeight))

    return result_img
````

## Standardize the output

With each loaded image, we also specify the expected output. For this, we use **one-hot encoding**.

* One-hot encode the labels. To do this, create an array of zeros representing each class of traffic light (red, yellow, green), and set the index of the expected class number to 1. 

Since we have three classes (red, yellow, and green), we have imposed an order of: [red value, yellow value, green value]. To one-hot encode, say, a yellow light, we would first initialize an array to [0, 0, 0] and change the middle value (the yellow value) to 1: [0, 1, 0].

### (IMPLEMENTATION): Implement one-hot encoding

````python
## TODO: One hot encode an image label
## Given a label - "red", "green", or "yellow" - return a one-hot encoded label

# Examples: 
# one_hot_encode("red") should return: [1, 0, 0]
# one_hot_encode("yellow") should return: [0, 1, 0]
# one_hot_encode("green") should return: [0, 0, 1]

def one_hot_encode(label):
    
    # Handle error when given label is not one of the following options
    if label != "red" and label != "green" and label != "yellow":
        raise Exception('Unknown label for error "one_hot_encode"-function:', str(label))
    
    ## TODO: Create a one-hot encoded label that works for all classes of traffic lights
    one_hot_encoded = {
        "red": [1, 0, 0],
        "yellow": [0, 1, 0],
        "green": [0, 0, 1]
    }
        
    return one_hot_encoded[label]
````

## Construct a `STANDARDIZED_LIST` of input images and output labels.

This function takes in a list of image-label pairs and outputs a **standardized** list of resized images and one-hot encoded labels.

This uses the functions you defined above to standardize the input and output, so those functions must be complete for this standardization to work!


````python
def standardize(image_list):
    
    # Empty image data array
    standard_list = []

    # Iterate through all the image-label pairs
    for item in image_list:
        image = item[0]
        label = item[1]

        # Standardize the image
        standardized_im = standardize_input(image)

        # One-hot encode the label
        one_hot_label = one_hot_encode(label)    

        # Append the image, and it's one hot encoded label to the full, processed list of image data 
        standard_list.append((standardized_im, one_hot_label))
        
    return standard_list

# Standardize all training images
STANDARDIZED_LIST = standardize(IMAGE_LIST)
````

### (IMPLEMENTATION): Create a brightness feature that uses HSV color space

Write a function that takes in an RGB image and returns a 1D feature vector and/or single value that will help classify an image of a traffic light. The only requirement is that this function should apply an HSV colorspace transformation, the rest is up to you. 

From this feature, you should be able to estimate an image's label and classify it as either a red, green, or yellow traffic light. You may also define helper functions if they simplify your code.

````python
def avg_threshold_image(image):
    avg = np.sum(image) / image.size
    threshold_map = image > avg
    image[threshold_map != True] = 0   # image[not c] does not work, so i use "!= True"
    return image
````

````python
def create_feature(rgb_image):
    
    # Convert RGB image to HSV
    hsv_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2HSV)
    
    # Get h, s, v, representation of the image separatly
    h_values = hsv_image[:,:,0]
    s_values = hsv_image[:,:,1]
    v_values = hsv_image[:,:,2]
    
    # Only keep values of the image (hsv v-representation) which have a 
    # value higher than the average all other values are set to 0
    masked_v_values = avg_threshold_image(v_values)
    # masked_s_values = avg_threshold_image(s_values)
    
    # Multiply saturation and value of hsv-image for analytics purpose
    # combined_s_v = hsv_image[:,:,1] * hsv_image[:,:,2]
    
    # Calc a third of the image height (len(masked_v_values) is equal to the image-height)
    # I use this instead of rgb_image.shape, because the calculation below bases on masked_v_values too
    third_length = round(len(masked_v_values) / 3)
    # Split the image vertical into three nearly same sized arrays (depends if the image-height can be dividied by 3)
    # The first part would most likely contain the traffic-lights red-part
    # The second part would most likely contain the traffic-lights yellow-part
    # The third part would most likely contain the traffic-lights green-part
    red_part = masked_v_values[:third_length]                           # Range 0/3 - 1/3
    yellow_part = masked_v_values[third_length:third_length * 2]        # Range 1/3 - 2/3
    green_part = masked_v_values[len(masked_v_values) - third_length:]    # Range 2/3 - 3/3
    
    # Sum the top, middle and bottom of the image v-values up to a single integer
    sum_green = np.sum(green_part)   
    sum_yellow = np.sum(yellow_part)
    sum_red = np.sum(red_part)
    
    # The following code finds the index of the highest value of the splitted parts of 
    # the image and set the specific position in the  one_hot_encode array to one.
    sum_array = [sum_red, sum_yellow, sum_green]
    result = [0, 0, 0]                                     # [red, yellow, green]
    index_with_max_value = sum_array.index(max(sum_array)) # get index of max value
    result[index_with_max_value] = 1                       # set value at specific index to one
    
    return result
````

## Determine the Accuracy

Compare the output of your classification algorithm (a.k.a. your "model") with the true labels and determine the accuracy.

This code stores all the misclassified images, their predicted labels, and their true labels, in a list called `MISCLASSIFIED`. This code is used for testing and *should not be changed*.

````python
# Constructs a list of misclassified images given a list of test images and their labels
# This will throw an AssertionError if labels are not standardized (one-hot encoded)

def get_misclassified_images(test_images):
    # Track misclassified images by placing them into a list
    misclassified_images_labels = []

    # Iterate through all the test images
    # Classify each image and compare to the true label
    for image in test_images:

        # Get true data
        im = image[0]
        true_label = image[1]
        assert(len(true_label) == 3), "The true_label is not the expected length (3)."

        # Get predicted label from your classifier
        predicted_label = estimate_label(im)
        assert(len(predicted_label) == 3), "The predicted_label is not the expected length (3)."

        # Compare true and predicted labels 
        if(predicted_label != true_label):
            # If these labels are not equal, the image has been misclassified
            misclassified_images_labels.append((im, predicted_label, true_label))
            
    # Return the list of misclassified [image, predicted_label, true_label] values
    return misclassified_images_labels


# Find all misclassified images in a given test set
MISCLASSIFIED = get_misclassified_images(STANDARDIZED_TEST_LIST)

# Accuracy calculations
total = len(STANDARDIZED_TEST_LIST)
num_correct = total - len(MISCLASSIFIED)
accuracy = num_correct/total

print('Accuracy: ' + str(accuracy))
print("Number of misclassified images = " + str(len(MISCLASSIFIED)) +' out of '+ str(total))
````

# 6. Analytics
The following code is very similar to the "create_feature"-function code above.
I've just added additionally functionality to visualize the images using only the h,s,v representation of the hsv image or combine those values.
The different visualization were really helpful to find new ideas an inspiration how to improve the algorithm.
Within this help i was able to improve my classification algorithm to be more accurate than 95%

Now the code visualize some of the missclassified images.

````python
def create_feature_analyse(rgb_image):
    # Convert RGB image to HSV
    hsv_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2HSV)
    
    # Get h, s, v, representation of the image separatly
    h_values = hsv_image[:,:,0]
    s_values = hsv_image[:,:,1]
    v_values = hsv_image[:,:,2]
    
    # Only keep values of the image (hsv v-representation) which have a 
    # value higher than the average all other values are set to 0
    masked_v_values = avg_threshold_image(v_values)
    masked_s_values = avg_threshold_image(s_values)
    
    fig, ((p0, p1, p2, p3), (p4, p5, p6, p7), (p8, p9, p10, p11)) = plt.subplots(3, 4, figsize=(30,20))
    
    # -----------------------------------
    # First row of subplots. (p0 - p3)
    # -----------------------------------
    # Original image as HSV representation
    p0.set_title("HSV-Image grayscaled", fontsize=20)
    p0.imshow(rgb_image)
        
    # H-value (hue) of the HSV-image representated in grayscale
    p1.set_title("Hue grayscaled", fontsize=20)
    p1.imshow(h_values, cmap='gray')
    
    # S-value (saturation) of the HSV-image representated in grayscale
    p2.set_title("Saturation grayscaled", fontsize=20)
    p2.imshow(s_values, cmap='gray')
    
    # V-value (value) of the HSV-image representated in grayscale
    p3.set_title("Value grayscaled", fontsize=20)
    p3.imshow(v_values, cmap='gray')
    
    # -----------------------------------
    # Seconds row of subplots. (p0 - p3)
    # -----------------------------------
    # Sum each row of the hsv, h-, s-, v-representation of the 
    # image separatly for analytic-/visualization-purpose
    sums_hsv_values = [np.sum(subarr) for subarr in hsv_image]
    sums_h_values = [np.sum(subarr) for subarr in h_values]
    sums_s_values = [np.sum(subarr) for subarr in s_values]
    sums_v_values = [np.sum(subarr) for subarr in v_values]
    
    # Show entire HSV-image (pixels-summed by rows) as horizontal barchart.
    p4.set_title("HSV (summed rows)", fontsize=20)
    p4.barh(range(0, len(sums_hsv_values)), sums_hsv_values)
    
    # Show hue (h) of HSV-image (pixels-summed by rows) as horizontal barchart.
    p5.set_title("Hue (summed rows)", fontsize=20)
    p5.barh(range(0, len(sums_h_values)), sums_h_values)
    
    # Show saturation (s) of HSV-image (pixels-summed by rows) as horizontal barchart.
    p6.set_title("Saturation (summed rows)", fontsize=20)
    p6.barh(range(0, len(sums_s_values)), sums_s_values)
    
    # Show value (v) of HSV-image (pixels-summed by rows) as horizontal barchart.
    p7.set_title("Value (summed rows)", fontsize=20)
    p7.barh(range(0, len(sums_v_values)), sums_v_values)

    
    # Multiply saturation and value of hsv-image for analytics purpose
    combined_s_v = hsv_image[:,:,1] * hsv_image[:,:,2]
    sums_sv_values = [np.sum(subarr) for subarr in combined_s_v]
    
    # Show value (v) multiplied with saturation (s) of HSV-image (pixels-summed by rows) as horizontal barchart.
    p8.barh(range(0, len(sums_sv_values)), sums_sv_values)
    
    # Show images saturation multiplied with value as grayscaled image
    p9.imshow(combined_s_v, cmap='gray')
    p10.imshow(rgb_image)
    
    # Invert y axis of value representation, this ist required because for charts
    # x0 y0 position would be bottom-left, for images x0 y0 is top-left.
    # By inverting the yaxis the chart is more likley representing the visualization of the image
    p4.invert_yaxis()
    p5.invert_yaxis()
    p6.invert_yaxis()
    p7.invert_yaxis()    
    p8.invert_yaxis()
    
    # Calc a third of the image height (len(masked_v_values) is equal to the image-height)
    # I use this instead of rgb_image.shape, because the calculation below bases on masked_v_values too
    third_length = round(len(masked_v_values) / 3)
    # Split the image vertical into three nearly same sized arrays (depends if the image-height can be dividied by 3)
    # The first part would most likely contain the traffic-lights red-part
    # The second part would most likely contain the traffic-lights yellow-part
    # The third part would most likely contain the traffic-lights green-part
    red_part = masked_v_values[:third_length]                           # Range 0/3 - 1/3
    yellow_part = masked_v_values[third_length:third_length * 2]        # Range 1/3 - 2/3
    green_part = masked_v_values[len(masked_v_values) - third_length:]    # Range 2/3 - 3/3
    
    # Sum the top, middle and bottom of the image v-values up to a single integer
    sum_green = np.sum(green_part)   
    sum_yellow = np.sum(yellow_part)
    sum_red = np.sum(red_part)
    
    # The following code finds the index of the highest value of the splitted parts of 
    # the image and set the specific position in the  one_hot_encode array to one.
    sum_array = [sum_red, sum_yellow, sum_green]
    result = [0, 0, 0]                                     # [red, yellow, green]
    index_with_max_value = sum_array.index(max(sum_array)) # get index of max value
    result[index_with_max_value] = 1                       # set value at specific index to one
    
    return result


images_to_analyse = MISCLASSIFIED[:5]
for image in images_to_analyse:
    create_feature_analyse(image[0])
````
